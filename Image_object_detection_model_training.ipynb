# Loading pretrained model and training it on coco8 dataset

```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolov8n.pt")  # load a pretrained model (recommended for training)

# Train the model
results = model.train(data="coco8.yaml", epochs=100, imgsz=640)
```

# Model validation across all categories

```python
# Validate the model
metrics = model.val()

# Print various mAP metrics
print("Overall mAP (IoU 0.50-0.95):", metrics.box.map)
print("mAP at IoU 0.50:", metrics.box.map50)
print("mAP at IoU 0.75:", metrics.box.map75)

# Print mAP for each category
category_maps = metrics.box.maps
for i, category_map in enumerate(category_maps):
    print(f"Category {i} mAP (IoU 0.50-0.95): {category_map}")
```

# Performing inference

```python
results = model("https://ultralytics.com/images/bus.jpg")
```

```python
# Extract the detected class indices
detected_class_indices = results[0].boxes.cls

# Extract the class names using the model's class names
detected_class_names = [model.names[int(idx)] for idx in detected_class_indices]

# Print the detected class names
print("Detected classes:")
for class_name in detected_class_names:
    print(class_name)
```

```python
print(detected_class_names)
```

```python
category_map = {}
for obj in detected_class_names:
    if obj in category_map:
        category_map[obj] += 1
    else:
        category_map[obj] = 1

print(category_map)
```

```python
print("Objects present in image:")
for key in category_map:
    print(key + " - count: " + str(category_map[key]))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
import requests
from io import BytesIO
import numpy as np

# Perform inference on the image
image_url = "https://ultralytics.com/images/bus.jpg"
response = requests.get(image_url)
img = Image.open(BytesIO(response.content))

# Convert PIL Image to OpenCV format
opencv_image = cv2.cvtColor(np.array(img), cv2.COLOR_RGB2BGR)

# Run the model on the image
results = model(opencv_image)

# Extract information from the results
for result in results:
    boxes = result.boxes.xyxy.cpu().numpy()  # Bounding boxes
    class_ids = result.boxes.cls.cpu().numpy()  # Class IDs
    confidences = result.boxes.conf.cpu().numpy()  # Confidence scores
    class_names = result.names  # Class names dictionary

    # Draw the bounding boxes and labels on the image
    for box, cls_id, confidence in zip(boxes, class_ids, confidences):
        x1, y1, x2, y2 = map(int, box)
        class_name = class_names[int(cls_id)]
        label = f"{class_name} {confidence:.2f}"

        # Draw the rectangle
        cv2.rectangle(opencv_image, (x1, y1), (x2, y2), (0, 0, 255), 2)

        # Put the label near the bounding box
        cv2.putText(opencv_image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2)

# Convert back to RGB for displaying with matplotlib
rgb_image = cv2.cvtColor(opencv_image, cv2.COLOR_BGR2RGB)

# Display the image with matplotlib
plt.figure(figsize=(10, 10))
plt.imshow(rgb_image)
plt.axis('off')
plt.show()
```
